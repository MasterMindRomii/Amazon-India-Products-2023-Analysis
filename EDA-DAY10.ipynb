# Data Visualization

```python
# Visualization 1: Histogram of 'stars' column
plt.figure(figsize=(8, 6))
plt.hist(df['stars'], bins=10, edgecolor='k')
plt.title('Distribution of Stars')
plt.xlabel('Stars')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 2: Bar plot of 'categoryName' counts
category_counts = df['categoryName'].value_counts()
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar')
plt.title('Product Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 3: Scatter plot of 'price' vs 'reviews'
plt.figure(figsize=(8, 6))
plt.scatter(df['price'], df['reviews'], alpha=0.5)
plt.title('Price vs. Number of Reviews')
plt.xlabel('Price')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
# Visualization 4: Box plot of 'price' by 'isBestSeller'
plt.figure(figsize=(8, 6))
df.boxplot(column='price', by='isBestSeller')
plt.title('Price Distribution by Best Seller')
plt.suptitle('')  # Remove default title
plt.xlabel('Best Seller')
plt.ylabel('Price')
plt.xticks([1, 2], ['Not Best Seller', 'Best Seller'])
plt.show()
```

```python
# Visualization 5: Count plot of 'boughtInLastMonth'
plt.figure(figsize=(6, 4))
df['boughtInLastMonth'].value_counts().plot(kind='bar')
plt.title('Bought in Last Month (0: No, 1: Yes)')
plt.xlabel('Bought in Last Month')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'], rotation=0)
plt.show()
```
